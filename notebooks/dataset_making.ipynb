This script was used to copy files from hard drive to a local computer on a specific location

```python
import os
import sys
from shutil import copy

images_dir = r'E:/SKU-110/SKU110K_fixed/images/'
up_level_dir = "./"
dest_dirs = ["train", 'val', 'test']

for dir in dest_dirs:
    os.makedirs(os.path.join(up_level_dir, dir), exist_ok=True)

if os.path.exists(images_dir):
    print(f'{images_dir} exists')

    for file in os.listdir(images_dir):
        src = os.path.join(images_dir, file)
        for dirname in dest_dirs:
            if file.startswith(dirname):
                dest = os.path.join(up_level_dir, dirname, file)
                continue
        if not os.path.exists(dest):
            copy(src, dest)
            print(f"{src} copied to {dest}")
```
